```python
#reads in image and outputs dataframe for clusters and pixels
import sys
import math
import timeit
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from skimage import io

#custom modules
import find_closest_cluster
import plot_rgb
import k_means
import winner_take_all
import fig_io

def run_k_means(n_clusters,pixel_data):
    count_max=10
    clusters= fig_io.initialize_clusters(n_clusters)
    clusters=k_means.initialize(clusters.copy())
    pixel_data_comp=pixel_data   
    counter=0
    limit_reached=False
    while(limit_reached==False):
        counter+=1
        pixel_data_comp=pixel_data
        
        #reset
        clusters=k_means.reset_clusters(clusters.copy())
        clusters.reset_index(drop=True,inplace=True)
        pixel_data=find_closest_cluster.reset_closest_cluster(pixel_data.copy())
        pixel_data.reset_index(drop=True,inplace=True)
        
        #algorithm
        pixel_data=find_closest_cluster.find_all_closest_cluster(pixel_data.copy(),clusters.copy())
        clusters=k_means.sample_mean(pixel_data,clusters.copy())
        clusters=k_means.reassign_cluster_pos(clusters.copy())
        
        #stop conditions
        if pixel_data_comp.equals(pixel_data)==True:
            limit_reached=True
            print('algorithm has converged stopping....')   
        print('counter='+str(counter))
        if counter==count_max:
            limit_reached=True
            print('algorithm has reached '+str(counter)+' iterations stopping....') 

            
    print('completed for '+str(n_clusters)+' clusters')
    pixel_data=find_closest_cluster.find_all_closest_cluster(pixel_data.copy(),clusters.copy())
    pixel_data=find_closest_cluster.update_pixels_to_closest_cluster(pixel_data.copy(),clusters.copy())
    
    plot_rgb.plot_rgb_cluster_pix(clusters.copy(),pixel_data,'k_means_'+str(n_clusters))
    plot_rgb.save_file(clusters.copy(),pixel_data.copy(),'k_means_'+str(n_clusters))

    
    
    
def run_winner_take_all(n_clusters,pixel_data,learning_parameter):
    count_max=5
    clusters=fig_io.initialize_clusters(n_clusters)
    clusters=winner_take_all.initialize(clusters.copy())
    pixel_data_comp=pixel_data   
    counter=0
    limit_reached=False
    
    while(limit_reached==False):
        counter+=1
        pixel_data_comp=pixel_data
        
        #reset
        clusters=winner_take_all.reset_clusters(clusters.copy())
        clusters.reset_index(drop=True,inplace=True)
        pixel_data=find_closest_cluster.reset_closest_cluster(pixel_data.copy())
        pixel_data.reset_index(drop=True,inplace=True)
        
        #algorithm
        pixel_data=find_closest_cluster.find_all_closest_cluster(pixel_data.copy(),clusters.copy())
        clusters=winner_take_all.pixel_shift(pixel_data.copy(),clusters,learning_parameter)
        
        #stop conditions
        if pixel_data_comp.equals(pixel_data)==True:
            limit_reached=True
            print('algorithm has converged stopping....')   
        print('counter='+str(counter))
        if counter==count_max:
            limit_reached=True
            print('algorithm has reached '+str(counter)+' iterations stopping....') 
        
    print('completed for '+str(n_clusters)+' clusters')
    pixel_data=find_closest_cluster.find_all_closest_cluster(pixel_data.copy(),clusters.copy())
    pixel_data=find_closest_cluster.update_pixels_to_closest_cluster(pixel_data.copy(),clusters.copy())
    
    plot_rgb.plot_rgb_cluster_pix(clusters.copy(),pixel_data,'wta_'+str(n_clusters))
    plot_rgb.save_file(clusters.copy(),pixel_data.copy(),'wta_'+str(n_clusters))
    
    
    
    
def run_kohonen(n_clusters,pixel_data,eps_max,eps_min):    
    count_max=5
    clusters=fig_io.initialize_clusters(n_clusters)
    clusters=winner_take_all.initialize(clusters.copy())
    pixel_data_comp=pixel_data   
    counter=0
    limit_reached=False
    
    while(limit_reached==False):
        counter+=1
        pixel_data_comp=pixel_data
        
        #reset
        clusters=winner_take_all.reset_clusters(clusters.copy())
        clusters.reset_index(drop=True,inplace=True)
        pixel_data=find_closest_cluster.reset_closest_cluster(pixel_data.copy())
        pixel_data.reset_index(drop=True,inplace=True)
        
        #algorithm
        pixel_data=find_closest_cluster.find_all_closest_cluster(pixel_data.copy(),clusters.copy())
        winner_take_all.kohonen(pixel_data=pixel_data.copy(),clusters=clusters.copy(),epsilon_max=eps_max,epsilon_min=eps_min,k=counter,k_max=count_max)
        
        #stop conditions
        print('counter='+str(counter))
        if counter==count_max:
            limit_reached=True
            print('algorithm has reached '+str(counter)+' iterations stopping....') 
        
    print('completed for '+str(n_clusters)+' clusters')
    pixel_data=find_closest_cluster.find_all_closest_cluster(pixel_data.copy(),clusters.copy())
    pixel_data=find_closest_cluster.update_pixels_to_closest_cluster(pixel_data.copy(),clusters.copy())
    
    plot_rgb.plot_rgb_cluster_pix(clusters.copy(),pixel_data,'kohonen_'+str(n_clusters))
    plot_rgb.save_file(clusters.copy(),pixel_data.copy(),'kohonen_'+str(n_clusters))
    
    
    
pixel_data=fig_io.read_in_fig('flowersm.ppm')
for n_clusters in [256,128,64,32,16,8,4,2]:
   # run_k_means(n_clusters,pixel_data)
    #run_winner_take_all(n_clusters,pixel_data,0.05)
    run_kohonen(n_clusters,pixel_data,2,0.5)
    
```
